```python
import os
import json

import time
import random
from datetime import datetime
import numpy as np
import pandas as pd
import folium
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def add_traj(trajectory_data, weight = 4, color = 'red', opacity = 0.8, attr='default', m = None, zoom_start = 14):
    # 提取经度纬度列表
    longitude_list, latitude_list = trajectory_data

    # 计算地图中心点
    center_longitude = sum(longitude_list) / len(longitude_list)
    center_latitude = sum(latitude_list) / len(latitude_list)

    # 创建地图对象
    # m = folium.Map(location=[center_latitude, center_longitude], zoom_start=13, attr=attr)
    if m is None:

        # m = folium.Map(location=[latitude_list[0], longitude_list[0]], zoom_start=15, attr=attr)
        m = folium.Map(location=[center_latitude, center_longitude], zoom_start=zoom_start, attr=attr)

    # 在地图上添加轨迹点
    # for lat, lon in zip(latitude_list, longitude_list):
    #     folium.PolyLine(location=[lat, lon], weight = weight, color = color, opacity = opacity).add_to(m)
    # return m
    folium.PolyLine(list(zip(latitude_list, longitude_list)), weight = weight, color = color, opacity = opacity).add_to(m)
    return m
```

```python
def sample_traj(filename):
    df = pd.read_csv(filename, delimiter=',')

    # 获取不同的entity_id
    unique_entities = df['traj_id'].unique()

    # 随机选择一个entity_id
    entity_id = random.choice(unique_entities)

    # 获取该entity_id对应的所有轨迹点
    entity_traj = df[df['traj_id'] == entity_id]

    # 提取经度纬度列表
    longitude_list = []#entity_traj['longitude'].tolist()
    latitude_list = [] #entity_traj['latitude'].tolist()

    for coor in entity_traj['coordinates'].tolist():
        longitude_list.append(float(coor.split(',')[0].strip("\"[]")))
        latitude_list.append(float(coor.split(',')[1].strip("\"[]")))

    return [longitude_list, latitude_list]
```

```python
filename = '../data/tul/chengdu/raw/chengdu_0304_u100.dyna'
traj = sample_traj(filename)
m1 = add_traj(traj, weight = 4, color = 'red', opacity = 0.8)
filename1 = '../data/tul/chengdu/raw/Chengdu_Sample13.dyna'
traj1 = sample_traj(filename1)
m2 = add_traj(traj1, weight = 4, color = 'blue', opacity = 0.8, m = m1)
```
