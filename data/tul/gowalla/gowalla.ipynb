```python
import pandas as pd
import matplotlib.pyplot as plt

# 使用pandas读取数据



csv_path = './raw/gowalla-all.csv'
data = pd.read_csv(csv_path, sep='\t')

traj_counts = data.drop_duplicates(subset=['ObjectID', 'TrajNumber']).groupby('ObjectID').size()

plt.bar(traj_counts.index.astype(str), traj_counts.values)
plt.xlabel('User ID')
plt.ylabel('Number of Trajectories')
plt.title('Number of Trajectories per User')
plt.tight_layout()
plt.show()
```
